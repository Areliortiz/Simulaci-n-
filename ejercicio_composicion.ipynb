```python
import random as rnd
import matplotlib.pyplot as plt
import numpy as np
```

Generar una muestra de la variable con ditribución.




```python
X= np.linspace(2,3, 50)
Y = X/2 - 1  #primer trozo de la función

V = np.linspace(3,6, 50)
W = -V/6 +1 #segundo trozo de la función

plt.figure()
plt.plot(X, Y)   #se grafican ambas funciones por partes (son dos partes)
plt.plot(V, W)

plt.grid()
plt.show()
```

2.$\int_{-\infty}^{\infty} f(x) \, dx = 1$ \
$\int_{2}^{3} x/2 -1 \, dx +\int_{3}^{6} -x/6 +1 \, dx $\
$=1/4 +3/4 = 1$\
Por lo tanto si es un fdp

La función de la grafica es una f.d.p\
Se comprobó que la integral total da como resultado 1\
además la función siempre es positiva para todo X en el intervalo [2,3] Y [3,6]\
se aplica el método de composición para su simulación\

```python
def F_u(n) : #Devuelve una lista con los valores de la función inversa
 lista3=[]
 for i in range(n):
  u_1=rnd.random()#Generar dos números aleatorios U(0,1)
  u_2=rnd.random() #variable a trabajar en la condiciones:
  if u_1<=1/4: #intervalo de valores menores que 1/2
   x=2+2*np.sqrt(u_1)
   lista3.append(x)
  else: #intervalo de valores menores que 1/2
   x=(6-np.sqrt((-12*u_2)+9))
   lista3.append(x)

 return lista3
```

```python
C=F_u(100000)
```

```python
#El siguiente código muestra histograma/densidad
plt.hist(C,bins=50)
plt.show()
```

La función de distribución acumaltiva es: \
F(x) = $\int_{-\infty}^{\infty} f(s) \, ds $\
F(x)_1 = $\int_{2}^{x}x/2 -1 \, ds $=$ 1/4(x-2)^2 $\
F(x)_2= $\int_{3}^{x}-x/6 +1 \, ds $=$-9/4 + x- (x^2/12) $


```python
x_1 = np.linspace(2, 3, 50)
y_1 = 1/4*(x_1-2)**2 #esta es la función acumulativa del primer trozo de la función

x_2 = np.linspace(3,6, 50)
y_2 = 1/4+(-9/4 + x_2 - (x_2**2/12)) #esta es la función acumulativa del segundo trozo de la función
plt.figure()
plt.plot(x_1, y_1)
plt.plot(x_2, y_2)
plt.grid()
plt.show()
#se considera el area de f(x)_1 para obtener la siguiente funcion
```
