Generar una muestra de la variable con ditribuci√≥n.\
$ f(x) = 20x (1-x)^3$ si $ 0<x<1$.


```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#la grafica de fdp, es la siguiente
x = np.linspace(0, 1, 50)
y= (20*x)*((1-x)**3)
plt.plot(x,y)
plt.show()
```

La funci√≥n de la grafica es una f.d.p\
Se comprob√≥ que la integral total da como resultado 1\
adem√°s la funci√≥n siempre es positiva para todo X en el intervalo [0,1]\
se aplica el m√©todo de la aceptaci√≥n y rechazo para su simulaci√≥n\
-El numero maximo va a ser x= $1/4$\
$Nota$\
$f(x)/g(x)=c$\
tomamos g(x)=1\
si $x= 1/4$ entonces $c= 20(1/4)(3/4)^3=135/64$\
\
 Por lo que c = 135/64 y g(x)= 1

metodo de rechazo
1. Generar Y con densidad g
2. Generar un n√∫mero aleatorio U
3. si U $\leq f(y)/g(y)$, hacer x=y \
 En caso contario regreasar al paso 1

```python
#El siguiente c√≥digo muestra histograma

def funcion(n=100):
    list=[]
    for i in range(n):
        y=rd.uniform(0,1)
        U=rd.uniform(0,1)
        if ((20*y)*(1-y)**3)/(135/64)>=U: #condicion para regresar x, paso 3
         x=y
         list.append(x)
    return list
```

```python
hist=funcion(10000) #manda a llamar la funcion
plt.hist(hist,bins=500)
plt.show()
```

La funci√≥n de distribuci√≥n acumaltiva es: \
F(x) = $\int_{-\infty}^{\infty} f(s) \, ds $\
F(x) = $\int_{0}^{x}  \,20xs (1-s)^3$ ds $= x^2 (10-20x +15x^2-4x^3) $

```python
# grafica la funci√≥n acumulativa
X = np.linspace(1, 0, 50)
Y = X**2 *((10 - 20*X) +( 15*X**2) - (4* X**3))   #esta es la funci√≥n acumulativa de la funci√≥n 0<x<1
plt.figure()


plt.plot(X, Y )
plt.show()
```
