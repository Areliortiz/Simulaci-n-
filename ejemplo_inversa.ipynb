```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def triangulo(n=100):
    list=[]
    for i in range(n):
        u=rnd.random()
        if 0<u<3/4:
          x=(np.sqrt((25/16)-(2*u))+5/4)
        elif 3/4<u<1:
          x=(4*u)+2
        else:
          x=0
        list.append(x)
    return list
```

```python
#usa el algoritmo
g= triangulo(100)
print(g)
```

```python
plt.plot(g)
#grafica, imprime valores de g
#sucesion
```
