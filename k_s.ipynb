```python
from tabulate import tabulate
datos=[0.8797,0.3884,0.6289,0.8750,0.5999,0.8589,0.9996,0.2415,0.3808,0.9606,0.9848,0.3469 ,0.7977, 0.5844, 0.8147, 0.6431, 0.7387, 0.5613, 0.0318, 0.7401,0.4557 ,0.1592, 0.8536, 0.8846, 0.3410, 0.1492 ,0.8681 ,0.5291, 0.3188, 0.5992 ,0.9170 ,0.2204 ,0.5991, 0.5461, 0.5739, 0.3254, 0.0856 ,0.2258 ,0.4603 ,0.5027,0.8376 ,0.6235 ,0.3681 ,0.2088 ,0.1525 ,0.2006 ,0.4720 ,0.4272 ,0.6360 ,0.0954]
#ordena de menor a mayor
datos.sort()

tabla=[]
i=0
n=len(datos)
Dmax=[]
dmax=[]

for xi in datos:
  i += 1
  #realiza operaciones para D+
  D=abs((i/n)-xi)
  Dmax.append(D)   #agraga los valores a una lista(Dmax)

  #realiza operaciones para D-
  d=abs(xi-((i-1)/n))
  dmax.append(d)    #agraga los valores a una lista(dDmenos)

  #crea tabla
  tabla.append([i,xi,D,d])
#imprime los maximos de D+ y de D-
maxD=max(Dmax)
print("el maximo D+ es ", maxD)
maxd=max(dmax)
print("el maximo D- es, ", maxd)

#imprime tabla
headers = ["i","Xi", "D+", "D-"]
print(tabulate(tabla, headers=headers, tablefmt="grid"))

#encuenta el valor maximo
if maxD >= maxd:
  print("El mayor es: ",maxD)
else:
  print("El mayor es: ",maxd)
```
