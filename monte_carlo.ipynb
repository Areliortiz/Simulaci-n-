```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def integral_monte_carlo(N):
    contador = 0
    lista = []
    for i in range(N):
        u1 = rnd.random()  # Genera un número aleatorio para x en el intervalo [0, 1]
        u2 = rnd.random()  # Genera un número aleatorio para y en el intervalo [0, 1]
        if u2 <= np.sqrt(np.arctan(u1)):  # Comprueba si el punto está debajo de la curva sqrt(arctan(x))
            contador += 1
        prob = contador / (i + 1)  # Calcula la probabilidad de que un punto esté debajo de la curva
        lista.append(prob)  # Agrega la probabilidad a la lista

    resultado_integral = lista[-1]  # El valor final de la lista es nuestra estimación de la integral
    return lista, resultado_integral
```

```python
# Llamamos a la función y graficamos los resultados
lista, resultado = integral_monte_carlo(1000)  # 1000 puntos de muestra
plt.plot(lista)
print("el valor del area bajo la curva es:",resultado)
# plt.axhline(resultado, color="red", label=f'Estimación de la integral: {resultado:.4f}')
plt.legend()
plt.show()
```

```python
# Función para realizar múltiples simulaciones
def simulacion(M):
    resultados = []
    for i in range(M):
        _, resultado = integral_monte_carlo(500)  # 500 puntos por simulación
        resultados.append(resultado)
    return resultados
```

```python
# Realizamos 1000 simulaciones y graficamos el histograma
resultados_simulacion = simulacion(1000)
plt.hist(resultados_simulacion, bins=30, edgecolor='black')
plt.title("Distribución de las estimaciones de la integral")
plt.show()
```
