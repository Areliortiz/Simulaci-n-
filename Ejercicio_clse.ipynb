Generar por el método de la transformada inversa.

$ f(x) = x^2/9$  si  $0 \leq x \leq 3$

comprobar que f(x) es fdp:\
I) $f(x) \geq 0$ en [0,3]\
   Siempre sera positivo ya que x se encuentra elevado al cuadrado.\
   por lo tanto cumple con la primera condición \
II)$\int_{-\infty}^{\infty} f(x) \, dx =1 $

  $\int_{0}^{3} x^2/9 \, dx = 3^3/27 = 1 $\
  cumple con la segunda restriccion.

   ### -> La función es una fdp ya que cumple con las dos restricciones

La función de distribución acumaltiva es: \
F(x) = $\int_{-\infty}^{\infty} f(s) \, ds $\
F(x) = $\int_{0}^{x} s^2/9 \, ds = x^3/27 $

Usamos el métod de la transformada inversa.\
sea U ~ u(0,1) y F(x)= U.\
$x^3/27 = u $\
Despejamos x\
${x = 3\sqrt[3]{u}} $


```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def inv(n=10):
    list=[]
    for i in range(n):
        u=rnd.random()
        x=3*pow(u,1/3)
        list.append(x)
    return list
```

```python
g= inv(100000)
#manda a llamar la funcion inv
```

```python
plt.hist(g,bins=50)
#representa la función densidad f(x)
```

```python
def cdf(g):
  n=len(g)
  x=np.sort(g) # ordena de manera ascendente
  y=np.arange(0,n)/n #  probabilidades acumulativas
  return x,y
```

```python
x,y=cdf(g)
```

```python
plt.step(x,y)
plt.show()
#Representa la funcion acumulativa F(x)
```
