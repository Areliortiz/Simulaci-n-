```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def inv(n=10):
    list=[]
    for i in range(n):
        u=rnd.random()
        x=3*pow(u,1/3)
        list.append(x)
    return list
```

```python
g= inv(100000)
print(g)
```

```python
plt.hist(g,bins=50)
```

```python
def cdf(g):
  n=len(g)
  x=np.sort(g)
  y=np.arange(0,n)/n
  return x,y
```

```python
x,y=cdf(g)
```

```python
plt.step(x,y)
#media
```
