```python
#progrma generador congruencial
from tabulate import tabulate
lista_principal = []
#pide valores para poder desarrollar el programa
m = float(input("dame el valor de m: "))
xo = float(input("dame el valor de la semilla (xo): "))
a = float(input("dame el valor de a: "))
c = float(input("dame el valor de c: "))
x=0

for x in range(30):
  #realiza operacion para encontrar nuevos valores y ralizar el for
  xnn=(a*xo)+c
  mod=xnn%m
  division=xo/m
  #genera tabla
  lista_principal.append([x, xo, xnn, mod, division])
  #cambia las variables para continuar con el for
  xo=mod
#imprime la tabla
headers = ["x", "xo", "xn+1", "mod","ui=xo/m"]
print(tabulate(lista_principal, headers=headers, tablefmt="grid"))
```
