```python
import random as rnd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def circulo(N):
  x=np.linspace (0,1,1000)
  n = 0
  for i in range(N):
    u1=rnd.random()
    u2=rnd.random()
    plt.scatter(u1,u2,color='red')
    if u2 <= np.sqrt(1-u1**2):
      n = n+ 1
  print("****",n)
  pi= 4*(n/N)
  print(pi)
  plt.plot(x,np.sqrt(1-x**2))
  plt.show(	)
  return pi
```
